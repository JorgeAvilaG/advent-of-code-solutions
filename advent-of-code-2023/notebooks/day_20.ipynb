```python
f = open("..\data\day_20.txt", "r")
data = list(f)
data = [line.rstrip() for line in data]
data[:5]
```

```python
# Data processing
def init_modules():
    pro_data = []
    for line in data:
        pro_data.append([line.split(' -> ')[0] , line.split(' -> ')[1].split(', ')])
    pro_data[:5]

    modules = {}
    conjunction = []
    for name, conections in pro_data:
        if name[0] == '%':
            modules[name[1:]] = ['%',conections, 0]
        elif name[0] == '&':
            modules[name[1:]] = ['&', conections, {}, []]
            conjunction.append(name[1:])
        elif name == 'broadcaster': # -> sr, ch, hd, bx'
            modules[name] = conections

    for name, conections in pro_data:
        for conect in conections:
            if conect in conjunction:
                modules[conect][3].append(0)
                modules[conect][2][name[1:]] = len(modules[conect][3])-1
    return modules



```

```python
###Part 1

def module_logic(pulse, module_name, origin):
    if module_name not in modules:
        return 1
    if module_name == 'broadcaster':
        for mod in modules[module_name]:
            to_check.append([pulse, mod, module_name])
        return 1
    if modules[module_name][0] == '%':
        if pulse == 1:
            return 1
        else:
            if modules[module_name][-1] == 0:
                new_pulse = 1
            else:
                new_pulse = 0
            modules[module_name][-1] = new_pulse
            for mod in modules[module_name][1]:
                to_check.append([new_pulse, mod, module_name])
            return 1
    else: # '&'
        position = modules[module_name][2][origin]
        modules[module_name][-1][position] = pulse
        if sum(modules[module_name][-1]) == len(modules[module_name][-1]):
            new_pulse = 0
        else:
            new_pulse = 1
        for mod in modules[module_name][1]:
            to_check.append([new_pulse, mod, module_name])
        return 1
    

from collections import deque 

low = 0
high = 0
modules = init_modules()

for n in range(1000):
    to_check = deque([[0,'broadcaster', 'origin']])

    while to_check:
        pulse, destiny, origin = to_check.popleft()
        if pulse == 0:
            low += 1
        else:
            high += 1

        module_logic(pulse, destiny, origin)
    #print(to_check)

print(low*high)
```
